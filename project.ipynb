Ensure both fields are correct, numeric, and meaningful

  --> Convert to numeric
  
  -->Remove invalid values (negative or missing)

```python
#convert to numeric
data['Cost'] = pd.to_numeric(data['Cost'], errors='coerce')
data['Sales'] = pd.to_numeric(data['Sales'], errors='coerce')
```

```python
#Remove invalid values (negative or missing) if it contains.
data = data[(data['Cost'] >=0) & (data['Sales'] >= 0)]
data=data.dropna(subset=['Cost','Sales'])
```

Remove Zero-Sales or Invalid Profit Records

```python
#create profit columns
data['Profit']=data['Sales']-data['Cost']
data['Profit']
```

```python
#remove zero sales
data=data[data['Sales']>0]
```

```python
#remove invalid profit record
data=data[data['Profit']>=0]
```

Profit Concentration (Pareto) Analysis

```python
#Sales Pareto
data=data.sort_values(by="Sales",ascending=False).reset_index(drop=True)
data["cum_sales"]=data["Sales"].cumsum()
data["cum_sales_pct"]=data["cum_sales"]/data["Sales"].sum()* 100
```

```python
#profit pareto
data = data.sort_values(by='Total_Profit', ascending=False).reset_index(drop=True)

data['cum_profit'] = data['Total_Profit'].cumsum()
data['cum_profit_pct'] = data['cum_profit'] / data['Total_Profit'].sum() * 100
```

```python
#sales 80%
rev_80 = data[data['cum_sales_pct'] <= 80]

rev_percent = len(rev_80) / len(data) * 100
print("Products contributing 80% revenue:", rev_percent, "%")
```

```python
#profit 80%
profit_80 = data[data['cum_profit_pct'] <= 80]

profit_percent = len(profit_80) / len(data) * 100
print("Products contributing 80% profit:", profit_percent, "%")
```

Revenue Pareto Chart


```python

# Step 2: Cumulative %
data['cum_sales_pct'] = data['Sales'].cumsum() / data['Sales'].sum() * 100

# Step 3: Plot
plt.figure()

# Bar chart (Seaborn)
sns.barplot(x=data.index, y='Sales', data=data)

# Line chart (Matplotlib)
plt.plot(data.index,data['cum_sales_pct'])

# 80% reference line
plt.axhline(y=80)
plt.xlabel("Products (Sorted Index)")
plt.ylabel("Sales / Cumulative %")
plt.title("Pareto Analysis - Revenue (Seaborn)")

plt.show()
```

Profit Pareto Chart


```python
sns.set_style("whitegrid")

# Step 1: Sort data by profit
data_sorted = data.sort_values(by='Total_Profit', ascending=False).reset_index(drop=True)

# Step 2: Calculate cumulative %
data_sorted['cum_profit_pct'] = data_sorted['Total_Profit'].cumsum() / data_sorted['Total_Profit'].sum() * 100

# Step 3: Plot
plt.figure()

# Bar chart
sns.barplot(x=data_sorted.index, y='Total_Profit', data=data_sorted)

# SINGLE smooth line
plt.plot(data_sorted.index, data_sorted['cum_profit_pct'])

# 80% reference line
plt.axhline(y=80)

plt.xlabel("Products (Sorted Index)")
plt.ylabel("Profit / Cumulative %")
plt.title("Pareto Analysis - Profit")

plt.show()
```
